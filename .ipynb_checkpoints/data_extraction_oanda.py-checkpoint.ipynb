```python
import numpy as np
```

```python
cols = data_df.columns[data_df.dtypes.eq(object)]
data_df[cols] = data_df[cols].apply(pd.to_numeric, errors='coerce', axis=0)
```

```python
data_df.info()
```

```python
data_df['returns'] = np.log(data_df['close'] / data_df['close'].shift(1))  # 12
data_df.info()
```

```python
cols = []
for momentum in [15, 30, 60, 120]:
    col = 'position_%s' % momentum
    data_df[col] = np.sign(data_df['returns'].rolling(momentum).mean())
    cols.append(col)
    
```

```python
import matplotlib

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns; sns.set()  # 18

strats = ['returns']  # 19

for col in cols:  # 20
    strat = 'strategy_%s' % col.split('_')[1]  # 21
    data_df[strat] = data_df[col].shift(1) * data_df['returns']  # 22
    strats.append(strat)  # 23

data_df[strats].dropna().cumsum().apply(np.exp).plot(figsize=(10,4)) # 24
plt.show()
```
